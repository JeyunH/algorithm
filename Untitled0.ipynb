```python
# 리스트와 반복문을 활용한 평균 구하기
students = ['철수', '영희', '민수', '지수', '준호']
scores = [85, 92, 78, 88, 90]
total = 0
sum_txt = '총합 = '

for i, score in enumerate(scores):
    total += score
#    sum_txt += students[i]+'('+str(score)+') + '
    sum_txt += f"{students[i]}({score})"
    if i+1 < len(scores):
        sum_txt += " + "

avg = total / len(scores)

print(sum_txt)
print(f"평균 = {total} / {len(scores)}명 = {avg}점")
```

```python
# while을 사용한 비밀 번호 비교 프로그램
correct_pwd = '1234'
tries = 3

while 0<tries:
    input_pwd = input("비밀번호를 입력하세요 : ")

    if input_pwd == correct_pwd:
        print("비밀번호 확인 완료! 서비스를 시작합니다")
        break

    tries -= 1
    print(f"비밀번호가 틀렸습니다. 남은 기회 : {tries}번")

if tries == 0 :
    print("비밀번호 3회 실패! 서비스 이용이 제한됩니다")
```

```python
# 기본적인 if-else 문
money = int(input("보유한 돈을 입력하세요 :"))
if money >= 10000:
    print("스테이크를 먹습니다!")
elif money >= 5000:
    print("피자를 먹습니다!")
elif money >= 3000:
    print("햄버거를 먹습니다!")
else:
    print("라면을 먹습니다!")
```

```python
# and 예제 (자바의 && 역할)
money = int(input("현재 가진 돈을 숫자로 입력하세요 (예: 6000) : "))
time = int(input("현재 시간을 입력하세요 (예: 14, 15 같은 숫자) : "))
is_event_time = 14<=time<=18

if money >= 5000 and is_event_time:
    print("1+1 행사 상품을 받습니다!")
else:
    print("1개만 정가로 구매합니다.")
```

```python
# 함수를 사용하여 숫자 계산 예제

def add_number(a,b):
    return a+b

print(add_number(5,7))

def square_number(n):
    return n*n

print(square_number(5))


print(square_number(add_number(5,7)))
```

```python
# 함수의 기본 매개변수 활용

def cook_ramen(water = 500, ramen = "신라면", soup = "매운맛 스프"):
    print(f"{water}ml 물을 끓입니다.")
    print(f"{ramen}을 넣습니다")
    print(f"{soup}을 넣습니다")
    print("라면 완성")

cook_ramen()
print("\n")
cook_ramen(600)
print("\n")
cook_ramen(600,"진라면")
print("\n")
cook_ramen(600,"진라면","순한맛 스프")
print("\n")
cook_ramen(600,None,"순한맛 스프")
```

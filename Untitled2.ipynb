```python
# 함수_ 키워드 매개변수 활용
def cook_ramen(water = 550, ramen="신라면", soup = "매운맛 스프"):
    print(f"{water}ml 물을 끓입니다")
    print(f"{ramen}을 넣습니다")
    print(f"{soup}을 넣습니다")
    print("라면이 완성되었습니다")

cook_ramen()
print("\n")
cook_ramen(600)
print("\n")
cook_ramen(600,"진라면")
print("\n")
cook_ramen(600,"진라면","순한맛 스프")
print("\n")
cook_ramen(soup="진한맛 스프", water=500, ramen="참깨라면")
print("\n")
```

```python
# 함수_ 가변 매개변수 활용
def 소개글(이름, 나이, *취미들):
    print(f"안녕하세요, 저는 {이름}입니다.")
    print(f"나이는 {나이}입니다")

    if 취미들:
        for 취미 in 취미들:
            print(f"- {취미}")
    else:
        print("취미는 아직 없습니다.")

    print()

소개글("철수", 25,"축구", "독서", "게임")
소개글("영희", 30,"요리", "등산")
소개글("민수", 22)
```

```python
cart = []

while True:
    print()
    print("===== 장바구니 프로그램 =====")
    print("1. 물건 추가")
    print("2. 물건 삭제")
    print("3. 장바구니 보기")
    print("4. 프로그램 종료")
    print()

    choice = input("원하는 기능의 번호를 입력하세요")

    if choice == "1":
        print()
        item = input("추가할 물건의 이름을 입력하세요: ")
        cart.append(item)
        print(f"{item}이(가) 장바구니에 추가되었습니다")
        print(f"현재 장바구니에는 {len(cart)}개의 물건의 있습니다")
    elif choice == "2":
        print()
        if not cart:
            print("장바구니가 비어있습니다")
        else:
            item = input("삭제할 물건의 이름을 입력하세요: ")
            if item in cart:
                cart.remove(item)
                print(f"{item}이(가) 장바구니에서 삭제되었습니다")
                print(f"현재 장바구니에는 {len(cart)}개의 물건의 있습니다")
            else:
                print(f"{item}은(는) 장바구니에 없습니다")
    elif choice == "3":
        print()
        if not cart:
            print("장바구니가 비어있습니다")
        else:
            print("장바구니에 있는 물건들:")
            for item in cart:
                print(f"- {item}")
            print(f"현재 장바구니에는 {len(cart)}개의 물건이 있습니다")
    elif choice == "4":
        print()
        print("프로그램을 종료합니다")
        break
    else:
        print("잘못된 입력입니다. 다시 시도해주세요")

```

```python
# 딕셔너리 생성하기
person = {
    "이름" : "홍길동",
    "나이" : 25,
    "직업" : "개발자"
}

print(person)

print(person["이름"])
# print(person["주소"]) -> 존재하지 않는 키를 호출하면 '키 에러' 발생
# 에러로 인한 프로그램 중단을 방지하려면 아래 처럼 get 메소드 사용
print(person.get("주소"))
# get 메소드를 사용할 때, 아래 처럼 None이 발생할 경우 반환하는 값을 선택할 수 있음
print(person.get("주소","존재하지 않는 Key입니다"))
```

```python
# 딕셔너리에 새로운 데이터 추가
person = {
    "이름" : "홍길동",
    "나이" : 25,
    "직업" : "개발자"
}
print(person)

person["주소"] = "쌍문동"
print(person)

# 딕셔너리에서 데이터 삭제
del person["주소"]
print(person)
```

```python
# 딕셔너리 반복문
scores = {
    "철수" : 85,
    "영희" : 92,
    "민수" : 78,
    "길동" : 81
}

for name in scores:
    print(name)

for score in scores.values():
    print(score)

for name, score in scores.items():
    print(f"{name}의 점수는 {score}입니다")
```

```python
# 이중 딕셔너리
others = {
    "김철수" : {"김치찌개" : 2, "된장찌개" : 1},
    "이영희" : {"된장찌개" : 3, "비빔밥" : 2}
}
for name, menus in others.items():
    print(f"{name}의 주문 목록은")
    for menu, qty in menus.items():
        print(f"- {menu} : {qty}개")
```

```python
# 리스트 안의 딕셔너리
from datetime import datetime
orders = [
    {"고객명" : "철수", "구매일" : "2024-01-01"},
    {"고객명" : "영희", "구매일" : "2023-10-10"}
]

cutoff = datetime.strptime("2024-03-01", "%Y-%m-%d")

inactive_customers = []

for order in orders:

    name = order["고객명"]

    date = datetime.strptime(order["구매일"], "%Y-%m-%d")

    if (cutoff - date).days > 90:
        inactive_customers.append(name)

print(inactive_customers)

```
